# NLP

```python
df['text'].replace({'r"\&\#[0-9]+\;"':''},regex=True,inplace=True)
# We removed the HTML characters
```

```python
df['text']=df['text'].str.lower()
# We turned it all into lowercase letters
```

```python
df['text'].replace({'[^\w\s]':''},regex=True,inplace=True)
# We removed the punctuation marks
```

```python
df['text'].replace({'\d+':''},regex=True,inplace=True)
# We removed the numbers
```

```python
df['text'].replace({'\n':'','\r':'',"\'":""},regex=True,inplace=True)
# Line break marks are removed
```

```python
df['text'].replace({'r"[\s]+"':''},regex=True,inplace=True)
# We removed unnecessary whitespaces
```

```python
from unicodedata import normalize
df['text']=df['text'].apply(lambda text: normalize("NFKD", str(text)).encode("ascii", "ignore").decode("utf-8", "ignore"))
# We removed the accents
```

```python
from nltk.corpus import stopwords
stop_words=stopwords.words('english')
df['text']=df['text'].apply(lambda x:" ".join([i for i in str(x).split(" ")  if i not in stop_words]))
# We removed STOPWORS 
```

# Language Detecting

```python
from langdetect import detect
df['language']=df['text'].apply(detect)
df=df[df['language']=='en']
df
```

# Tokenization

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
df['tokenized']=[tokenizer.tokenize(i) for i in df['text'].values]
```

```python
df.tokenized
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
df['lemmatize']=[[lemmatizer.lemmatize(str(i)) for i in words] for words in df['text'].values]
```

```python
df
```

```python
df.loc[df['polarity']>0.5,'sentiment']=2 #positive
df.loc[df['polarity']==0.5,'sentiment']=1 #neutral
df.loc[df['polarity']<0.5,'sentiment']=0 #negative
df
```

```python
import numpy as np
```

```python
import os
from PIL import Image
from wordcloud import WordCloud,STOPWORDS

def create_wordcloud(text):
    mask = np.array(Image.open("cloud.png"))
    stopwords = set(STOPWORDS)
    wc = WordCloud(background_color="white",
                  mask = mask,
                  max_words=25000,
                  stopwords=stopwords)
    wc.generate(str(text))
    wc.to_file("cloud3.png")
    print("Word Cloud Saved Successfully")
    path="cloud3.png"
    display(Image.open(path))
```

```python
create_wordcloud(df[df['sentiment']==2].text) # Positive
```

```python
create_wordcloud(df[df['sentiment']==0].text) # Negative
```
